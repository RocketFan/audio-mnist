# Plot Dataset distribution

```python
import os
import matplotlib.pyplot as plt

data_folder = 'data'
filenames = []

for root, dirs, files in os.walk(data_folder):
    for file in files:
        filenames.append(file)
```

```python
files_per_number = {}
files_per_person = {}

for file in filenames:
    params = file.split('_')
    if (params[0].isnumeric()):
        files_per_number[params[0]] = files_per_number.get(params[0], 0) + 1
    if (params[1].isnumeric()):
        files_per_person[params[1]] = files_per_person.get(params[1], 0) + 1
```

```python
# sort dictionaries
sorted_files_per_number = dict(sorted(files_per_number.items()))
sorted_files_per_person = dict(sorted(files_per_person.items()))
```

```python
# Plotting the number of files for each person
plt.figure(figsize=(10, 5))
plt.bar(sorted_files_per_person.keys(), files_per_person.values())
plt.xlabel('Person')
plt.ylabel('Number of Files')
plt.title('Number of Files for Each Person')
plt.xticks(rotation=90)
plt.show()

# Plotting the number of files for each number
plt.figure(figsize=(10, 5))
plt.bar(sorted_files_per_number.keys(), sorted_files_per_number.values())
plt.xlabel('Number')
plt.ylabel('Number of Files')
plt.title('Number of Files for Each Number')
plt.xticks(rotation=90)
plt.show()
```
